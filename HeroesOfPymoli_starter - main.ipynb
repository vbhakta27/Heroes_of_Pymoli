```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
# Get the length of a data frame with unique SNs
player_count = len(purchase_data["SN"].unique())

# Create new data frame with total count of unique SNs
total_players = pd.DataFrame({'Total Players': [player_count]})
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Get the length of a data frame with unique Item Names
unique_items_count = len(purchase_data["Item ID"].unique())

# Get the average price of the purchase data column and formats it
avg_price = "${:,.2f}".format(purchase_data['Price'].mean())

# Get the total count of purchases made
purchase_count = purchase_data["Purchase ID"].count()

# Sums up Price column for each transaction to get total revenue
total_revenue = "${:,.2f}".format(purchase_data['Price'].sum())

# Create new data frame with data
purchasing_analysis = pd.DataFrame({'Number of Unique Items': [unique_items_count],
                                   'Average Price': avg_price,
                                   'Number of Purchases': purchase_count,
                                    'Total Revenue': total_revenue})
purchasing_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Sets a dataframe that I can use for this code
age_df = unique_SN

# Create bins for age ranges 
bins = [0, 9, 14, 19, 24, 29, 34, 39, 150]

#Create bin labels showing ranges
bin_labels =['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

# Slices the data based on age into bins and 
# places the data series into a new column inside of the DataFrame
age_df["Age Range"] = pd.cut(age_df["Age"],bins, labels = bin_labels)

# Creates a Groupby object based on age range
age_group = age_df.groupby("Age Range")

# Stores the count of each age range into a list
age_counts = age_group["Age"].count().tolist()

# Loops throught to calculate percentage of age groups
age_percentages= [x *100/ player_count for x in age_counts]

# Create dataframe with data
age_demo = pd.DataFrame({'Age Ranges': bin_labels,
                                   'Total Count': age_counts,
                                   'Percentage of Players': age_percentages})
# Formats column into percentage
age_demo['Percentage of Players'] = age_demo['Percentage of Players'].map("{:.2f}%".format)

# Set age ranges as index
age_demo = age_demo.set_index('Age Ranges')
age_demo
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Sets a dataframe that I can use for this code
top_spenders = purchase_data

# Make a column with just 1 for all values
top_spenders["Purchase Count"]=1

# Groupby SN and take sum of all columns to get total paid by SN and purchase count total
top_spenders = top_spenders.groupby(["SN"]).sum()

# Renames column
top_spenders = top_spenders.rename(columns={"Price":"Total Purchase Value"})

# Sort table decending total paid
top_spenders = top_spenders.sort_values(by=['Total Purchase Value'],ascending=False)

# Calculates average price
top_spenders['Average Purchase Price'] = top_spenders['Total Purchase Value']/top_spenders['Purchase Count']

# Formats columns into dollars
top_spenders['Average Purchase Price'] = top_spenders['Average Purchase Price'].map("${:.2f}".format)
top_spenders['Total Purchase Value'] = top_spenders['Total Purchase Value'].map("${:.2f}".format)

# Shows all rows and the three columns we care about
top_spenders= top_spenders.loc[:, ["Purchase Count", "Average Purchase Price","Total Purchase Value"]]

# Shows the top five big spenders
top_spenders= top_spenders.iloc[0:5, :]
top_spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Sets a dataframe that I can use for this code
items_df = purchase_data

# Make a column with just 1 for all values
items_df["Purchase Count"]=1
items_df = items_df.groupby(["Item ID","Item Name"]).sum()

items_df = items_df.rename(columns={"Price":"Total Purchase Value"})
items_df["Item Price"] = items_df['Total Purchase Value'] / items_df['Purchase Count']

# Shows all rows and the three columns we care about
items_df = items_df.loc[:, ["Purchase Count", "Item Price","Total Purchase Value"]]

# Sorts by descending purchase count
popular_items = items_df.sort_values(by=['Purchase Count'],ascending=False)

# Formats columns into dollars
popular_items['Item Price'] = popular_items['Item Price'].map("${:.2f}".format)
popular_items['Total Purchase Value'] = popular_items['Total Purchase Value'].map("${:.2f}".format)

popular_items.head()
```
